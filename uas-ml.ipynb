```python
### Pemilihan fitur untuk model
#Mempertimbangkan hanya 2 fitur (Jumlah Toko dan Harga Unit)
X= data.iloc[:, [3,4]].values
```

```python
#Membangun Model Algoritma 
#KMeans untuk menentukan jumlah cluster yang optimal , KMeans++ menggunakan Elbow Mmethod
#untuk mengetahui K untuk KMeans, saya akan menggunakan Metode ELBOW pada Perhitungan KMEANS++
from sklearn.cluster import KMeans
wcss=[]

#kami selalu menganggap jumlah cluster maksimal adalah 10
#Anda dapat menilai jumlah cluster dengan melakukan rata-rata
###Kode statis untuk mendapatkan jumlah maksimum cluster

for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters= 5, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
